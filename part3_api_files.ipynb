```python
# Part B — Read:

# initialise line count to 0
line_count=0
# open file in read mode
with open("python_notes.txt", "r", encoding="utf-8") as file:
  # run a for loop to check each line in file
  for i,line in enumerate(file,start=1):
    # strip removes unnecessary \n characters
    clean_line=line.strip()
    print(f'{i}. {clean_line}')
    # increment the line counter after each line
    line_count+=1

print(f'\ntotal number of lines in the file: {line_count}')
```

```python
# Task 2 — API Integration

# Step 1 — Fetch and Display Products:

import requests
base_url="https://dummyjson.com/products"

# Send API request to fetch products
response=requests.get(f'{base_url}?limit=20')
# Convert API response to dictionary  using json
data=response.json()

# Extract list of products from the response
products=data['products']

print("\n---Fetched 20 Products ---\n")
print(f"{'ID':4} | {'Title':50} | {'Category':15} | {'Price':8} | {'Rating'}")
print("-" * 100)

for product in products:
  print(f'{product['id']:4} | {product['title']:50} | {product['category']:15} | ${product['price']:8} | {product['rating']}')
```

```python
# Step 2 — Filter and Sort:

# function that returns product price
def get_price(p):
    return p["price"]

# creates a list of product dictionaries with product rating >= 4.5
filtered_products = [p for p in products if p["rating"] >= 4.5]

# sorts products in descending order
filtered_products.sort(key=get_price,reverse=True)

print('rating >=4.5 and sorted in desc price\n')
print(f"{'ID':4} | {'Title':50} | {'Category':15} | {'Price':8} | {'Rating'}")
print("-" * 100)
for product in filtered_products:
    print(f'{product['id']:4} | {product['title']:50} | {product['category']:15} | ${product['price']:8} | {product['rating']}')
```

```python
# Step 3 — Search by Category:- laptops

# Send API request to fetch products from category laptops
response=requests.get(f'{base_url}/category/laptops')
# Convert API response to dictionary  using json
data=response.json()

# Extract list of laptop products from the response
laptops=data['products']

print(f"{'Title':50} | {'Price'}")
print("-" * 70)
for product in laptops:
    print(f'{product['title']:50} | ${product['price']}')
```

```python
# Task 3 — Exception Handling

# Part A — Guarded Calculator:
def safe_divide(a, b):
  try:
    return a / b
  except ZeroDivisionError:
    return "Error: Cannot divide by zero"
  except TypeError:
    return "Error: Invalid input types"

# Test Cases
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

```python
# Part B — Guarded File Reader:

# initially set file found flag to be false
file_found=False
def read_file_safe(filename):
  try:
    with open(filename, "r", encoding="utf-8") as file:
      # if read file exists set flag to be true and return file content
      file_found=True
      content = file.read()
      return content

  except FileNotFoundError:
    # if file doesn't exist set flag to be false and return error message
    file_found=False
    return f"Error: File '{filename}' not found."

  finally:
    print(f"File operation attempt complete. - {'File Found' if file_found==True else 'File Not Found'}")

# Test Cases
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))

```

```python
# Part C — Robust API Calls:
import requests

# GET request example
try:
  response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
  data = response.json()
  products=data['products']

  print("\n---GET request example---\n")
  print(f"{'ID':4} | {'Title':50} | {'Category':15} | {'Price':8} | {'Rating'}")
  print("-" * 100)

  for product in products:
    print(f'{product['id']:4} | {product['title']:50} | {product['category']:15} | ${product['price']:8} | {product['rating']}')
except requests.exceptions.ConnectionError:
  print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
  print("Request timed out. Try again later.")
except Exception as e:
  print(f"Unexpected error: {e}")

# GET request by category example
try:
  response = requests.get("https://dummyjson.com/products/category/laptops", timeout=5)
  data = response.json()
  laptops=data['products']

  print('\n---GET request by category example---')
  print(f"\n{'Title':50} | {'Price'}")
  print("-" * 70)
  for product in laptops:
    print(f'{product['title']:50} | ${product['price']}')
except requests.exceptions.ConnectionError:
  print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
  print("Request timed out. Try again later.")
except Exception as e:
  print(f"Unexpected error: {e}")

# POST request example
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post("https://dummyjson.com/products/add",
        json=new_product,
        timeout=5
    )
    data = response.json()
    print('\n---POST request example---')
    print(data)
except requests.exceptions.ConnectionError:
  print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
  print("Request timed out. Try again later.")
except Exception as e:
  print(f"Unexpected error: {e}")
```

```python
# Part D — Input Validation Loop:

# initialise quit flag to 0
quit_flag=0

# while user doesn't type quit
while(quit_flag!=1):
  # strip the user input so that unnecessary blank spaces are removed
  product_id=input("Enter a product ID to look up (1–100), or 'quit' to exit: ").strip()

  # convert the input to lower to check if the user enters quit and break immediately
  if (product_id.lower()=='quit'):
    quit_flag=1
    break

  # check if the input entered is not int , if so go back to while loop to take another input
  if not product_id.isdigit():
    print("\nInvalid input. Please enter a number.")
    continue

  # check if the input entered is an int between 1-100 , if not go back to while loop
  if int(product_id) < 1 or int(product_id) > 100:
    print("\nPlease enter a number between 1 and 100.")
    continue

  # if input entered is int between 1-100
  try:
    # pass id through API to fetch product
    response = requests.get(f'https://dummyjson.com/products/{product_id}', timeout=5)

    if response.status_code==404:
      # if product doesn't exist
      print("Product not found.")
    elif response.status_code==200:
      # if product exists, return product data
      data = response.json()
      print(f"\n{data['title']} — ${data['price']}\n")

  except requests.exceptions.ConnectionError:
      print("Connection failed. Please check your internet.")

  except requests.exceptions.Timeout:
      print("Request timed out. Try again later.")

  except Exception as e:
      print(f"Unexpected error: {e}")

```

```python
# Task 4 — Logging to File

from datetime import datetime
import requests

# log error function
def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"

    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)

# test 1: connection error
try:
    bad_url = "https://this-host-does-not-exist-xyz.com/api"
    requests.get(bad_url, timeout=5)

except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError","No connection could be made")
    print("ConnectionError logged successfully.")

# test 2:  Trigger and log an HTTP-style error manually using status_code
try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    if response.status_code != 200:
      log_error("lookup_product", "HTTPError", f"{response.status_code} Not Found for product ID 999")
      print("HTTP error logged successfully.")

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", "ConnectionError","Please check your internet connection")
    print("ConnectionError logged successfully.")

except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out")
    print("Timeout logged successfully.")

except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))
    print("Unexpected error logged successfully.")

print("\n--- Error Log Contents ---\n")
with open("error_log.txt", "r", encoding="utf-8") as file:
    print(file.read())
```
